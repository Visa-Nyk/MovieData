```python
from collections import defaultdict
import unicodedata
from dateutil.parser import parse, ParserError
from datetime import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize'] = [10,10]
```

### Data loading
The data in this analysis is IMDb rating data for movies with no less than 10 000 ratings

```python
data = {}
for name in ["basics", "ratings", "crew", "release_info", "detailed_rating", "people", "principals"]:
    data[name] = pd.read_csv(f"{name}.csv.gz", compression = "gzip")
```

```python
for name in ["crew", "detailed_rating", "ratings", "release_info", "basics"]:
    if "tconst" in data[name].columns:
        data[name].set_index("tconst", inplace = True)
        data[name].sort_index(inplace = True)
```

```python
data["basics"].drop(columns = ["isAdult", "endYear"], inplace = True)
data["basics"]["genres"] = data["basics"]["genres"].apply(lambda s: s.split(","))
data["release_info"]["country"] = data["release_info"]["country"].apply(eval)
data["release_info"]["language"] = data["release_info"]["language"].apply(eval)
for col in data["crew"].columns:
    data["crew"][col] = data["crew"][col].apply(lambda s: s.split(","))
```

```python
data["release_info"]["release"] = pd.to_datetime(data["release_info"]["release"])
```

```python
names = pd.read_csv("names.csv")
## The names are ordered descending by frequency, so with gender neutral names the more frequent name is assumed
names.drop_duplicates("Name", inplace = True)
names.set_index("Name", inplace = True)
names.sort_index(inplace = True)
```

```python
def strip_accents(s):
    return ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')
def get_gender(name):
    first_name = name.split()[0]
    if first_name not in names.index:
        first_name = strip_accents(first_name)
        if first_name not in names.index:
            return "U"
    
    return names.loc[first_name, "Gender"]
```

```python
data["people"]["assumed_gender"] = data["people"]["primaryName"].apply(get_gender)
```

# Genrewise ratings
First we find out all the genres in the dataset:

```python
from functools import reduce
reduce(lambda x, y: set(x)|set(y), data["basics"]["genres"])
```

```python
def melt_like(df, by_variables, inc_variables=[], group_name="group"):
    res=pd.DataFrame()
    for var in by_variables:
        temp=df[[var]+inc_variables].copy()
        temp[group_name]=var[var.find("_")+1:]
        temp.rename(columns={var:var[:var.find("_")]},inplace=True)
        res=res.append(temp)
    return res
```

```python
genre_stats = data["detailed_rating"].join(data["basics"]["genres"])\
                                     .explode("genres", ignore_index = True)\
                                     .rename(columns = {"genres": "genre"})
genre_stats = genre_stats[genre_stats["genre"].isin({'Action',
                                                     'Animation',
                                                     'Comedy',
                                                     'Family',
                                                     'Musical',
                                                     'Romance',
                                                     'Fantasy',
                                                     'Sci-Fi',
                                                     'Horror',
                                                     'Documentary',
                                                     'Sport',
                                                     'History',
                                                     'War',
                                                     'Western'})]
genre_stats_by_sex = melt_like(genre_stats, ["rating_men_all_ages","rating_women_all_ages"], ["genre"], group_name="sex")
```

```python
bp = sns.barplot(x="genre", y="rating", hue="sex", data=genre_stats_by_sex)
bp.set_ylim((5, 8))
plt.xticks(rotation=90)
plt.tight_layout()
```

Women always rate movies a bit higher than men, but the difference is greatest in Musical, Family and Animation. The difference seems to be smallest in Western movies.

```python
genre_stats_by_age = melt_like(genre_stats, ["rating_all_under_29", "rating_all_30-45", "rating_all_over_45"], ["genre"], group_name="age group")
```

```python
bp = sns.barplot(x="genre", y="rating", hue="age group", data=genre_stats_by_age)
bp.set_ylim((5, 8))
plt.xticks(rotation=90)
plt.tight_layout()
```

Mostly follows the trend of decreasing rating with age. The only movies that the oldest age group rates higher than the middle age group are War and Western-movies. Under 18 age group merged with 18-29 as it has insufficient data.

# Data from finnish film institute
Box office, subsidy, release data from the Finnish Film Institute

```python
fmi = pd.read_csv("ses/ses.csv")
def try_parse(date_str):
    try:
        return parse(date_str)
    except ParserError:
        return np.nan
fmi["Ensi-ilta"] = fmi["Ensi-ilta"].apply(try_parse)
```

```python
fi = fmi[fmi["Tuotantomaa"].str.extract("(FI)").notna().loc[:, 0]]
```

### Inflation adjusted subsidies and box office-results 

```python
def get_values_grouped_by_year(ses_data,variable):
    res = pd.DataFrame(ses_data.groupby("Ensi-iltavuosi")[variable].sum())
    res["label"] = variable
    res.rename(columns = {variable: "value"},inplace=True)
    return res

subsidies=get_values_grouped_by_year(fi[fi["Ensi-iltavuosi"]>1985],"Tuotannon tuet_inf_adj")
b_o=get_values_grouped_by_year(fi[fi["Ensi-iltavuosi"]>1985],"Lipputulot_inf_adj")
sns.lineplot(x="Ensi-iltavuosi",y="value",hue="label",data=subsidies.append(b_o))
```

```python
def two_scale_plot(df,var,x_axel,plotting_function,group_name="variable",value_name="value"):
    scale=df[df[group_name] != var][value_name].mean()/df[df[group_name] == var][value_name].mean()
    df.loc[df[group_name] == var,value_name] = df[df[group_name] == var][value_name] * scale
    fig,ax1 = plt.subplots()
    plotting_function(x = x_axel, y = value_name, hue = group_name, data=df)
    ax2= ax1.twinx()
    ax2.set_ylim(ax1.get_ylim())
    ax2.set_yticklabels(np.round(ax1.get_yticks()/scale))
```

```python
fmi = fmi[fmi["Ensi-ilta"]>datetime(1800,1,1)]   
fmi = fmi[fmi["Ensi-iltakuu"]>0]
viewers = fmi.groupby(["Ensi-iltavuosi","Ensi-iltakuu"])["Katsojat per 100000"]. \
                                                          aggregate(["mean", "count"]). \
                                                          groupby("Ensi-iltakuu").mean()
viewers.rename(columns = {"mean": "viewers per 100 000", "count": "Movies Released"}, inplace = True)
viewers["month"] = viewers.index
viewers = viewers.melt(id_vars = "month")
```

## Number of movie viewers and movie premieres by month

```python
two_scale_plot(viewers, "Movies Released", "month", sns.barplot)
```

## Number of movie viewers and movie premieres by year

```python
viewers=pd.DataFrame(fmi.groupby("Ensi-iltavuosi")["Katsojat per 100000"].aggregate(["sum", "count"]))
viewers["year"]=viewers.index
viewers.rename(columns = {"sum": "Viewers by 100 000", "count": "Movies Released"}, inplace = True)

viewers = viewers.melt(id_vars = "year")
two_scale_plot(viewers,"Movies Released","year",sns.lineplot)        
```
